## Data Distribution

## Sales Distribution (Histogram)

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Sales'], kde=True, bins=40)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

## Profit Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Profit'], kde=True, bins=40, color='red')
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

## Discount Distribution

```python
plt.figure(figsize= (10, 5))
sns.histplot(df['Discount'], kde= True, bins = 20, color= 'green')
plt.title('Discount Distribution')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

## Quantity Distribution

```python
plt.bar(df['Quantity'].unique(), df['Quantity'].value_counts())
plt.title("Quantity Count")
plt.xlabel("Quantity")
plt.ylabel("Count")
plt.show()
```

## Category Distribution

## Exploratory Data Analysis (EDA)

## Sales vs Profit (Scatter Plot)

```python
plt.figure(figsize= (8, 5))
plt.scatter(df['Sales'], df['Profit'], label= 'Sales', alpha= 0.5)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.legend()
plt.show()
```

## Category vs Profit

```python
plt.figure(figsize= (8,5))
sns.boxplot(x= 'Category', y= 'Profit', data= df)
plt.title('Category Vs Profit')
plt.legend()
plt.show()
```

## Region vs Sales

```python
plt.figure(figsize= (8,5))
sns.barplot(x= 'Region', y= 'Sales', data= df)
plt.title('Region Vs Sales')
plt.show()
```

## Discount Vs Profit

```python
plt.figure(figsize= (8,5))
sns.scatterplot(x= 'Discount', y= 'Profit', data= df)
plt.title('Discount Vs Profit')
plt.grid(True)
plt.show()
```

## Sub - Category vs Sales

```python
plt.figure(figsize= (12,5))
sns.barplot(x= 'Sub-Category', y= 'Sales', data= df)
plt.title('Sub-Category Vs Sales')
plt.xticks(rotation= 45)
plt.show()
```

## Correlation Matrix (Numeric Columns)

```python
plt.figure(figsize= (6,4))
corr = df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
sns.heatmap(corr, annot= True, cmap= 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
